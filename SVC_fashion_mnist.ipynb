```python
import tensorflow
```

```python
from tensorflow.keras.datasets import fashion_mnist
import numpy as np
```

```python
(x_train,y_train),(x_test,y_test) = fashion_mnist.load_data()
```

```python
dic = {0:'T-shirt/top',
       1:'Trouser',
       2:'Pullover',
       3: 'Dress',
       4:'Coat',
       5:'Sandal',
       6:'Shirt',
       7:'Sneaker',
       8:'Bag',
       9:'Ankle boot'}
```

```python
x_train = x_train.reshape(60000,28*28)
x_test  = x_test.reshape(10000,28*28)
x_train = x_train/255
x_test  = x_test/255
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
estimator = SVC(probability = True,decision_function_shape= 'ovr')
C = np.array([1,5,10,20,50,100])
kernel = np.array(['poly','rbf'])
gamma  = np.array([0.0001,0.001,0.01,0.1,1,10])
degree = np.array([1,2,3])
param_grid = {'C':C,'kernel':kernel,'gamma':gamma,'degree':degree}
clt = GridSearchCV(estimator,param_grid,cv=5,return_train_score = True,scoring ='roc_auc',verbose=2 )
```

```python
clt.fit(x_train[0:250],y_train[0:250])
```

```python
clt.best_params_
```

```python
clf = SVC(probability=True,C=1,degree = 1,gamma= 0.0001,kernel = 'poly')
```

```python
clf.fit(x_train,y_train)
```

```python
import pickle
```

```python
saved_model_svc = pickle.dumps(clf)
```

```python
from joblib import Parallel, delayed
import joblib
```

```python
joblib.dump(clf, 'fashion_mnist_svc.pkl')
```

```python
y_pred = clf.predict(x_test)
```

```python
y_test.shape,y_pred.shape
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
